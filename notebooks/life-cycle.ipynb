```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    # if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
    #     count+=1
    #     continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)
```

```python
import logging, os, shutil, time
import mlflow
import numpy as np
import pandas as pd

from sklearn.model_selection    import train_test_split, GridSearchCV, KFold
from sklearn.compose            import ColumnTransformer
from sklearn.preprocessing      import OneHotEncoder, StandardScaler
from sklearn.pipeline           import Pipeline
from xgboost                    import XGBRegressor
from sklearn.metrics            import r2_score, mean_absolute_error, mean_squared_error


mlflow.set_experiment('XGBoost Regressor Hyperparameter Tuning')

# ─────────────────────────────────────────────────────────────────────────────
# 0. CONFIGURE LOGGING
# ─────────────────────────────────────────────────────────────────────────────
logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s %(message)s")

# ─────────────────────────────────────────────────────────────────────────────
# 1. LOAD YOUR DATA
# ─────────────────────────────────────────────────────────────────────────────
# assume df is your DataFrame and 'runs_y' is the target
# df = pd.read_csv("…")
X = final_df.drop(columns=['runs_y'])
y = final_df['runs_y']

# ─────────────────────────────────────────────────────────────────────────────
# 2. SPLIT TRAIN / TEST
# ─────────────────────────────────────────────────────────────────────────────
test_size = ts
random_state = 1
logging.info(f"Splitting data, test_size={test_size}")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=random_state
)

# ─────────────────────────────────────────────────────────────────────────────
# 3. DEFINE PREPROCESSOR + PIPELINE
# ─────────────────────────────────────────────────────────────────────────────
cat_cols = ['batting_team', 'bowling_team', 'city']
num_cols = ['current_score', 'balls_left', 'wickets_left', 'crr', 'last_five']

preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols),
    ('scale',  StandardScaler(),                      num_cols)
])

pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model',      XGBRegressor(
        objective='reg:squarederror',
        random_state=random_state
    ))
])

# ─────────────────────────────────────────────────────────────────────────────
# 4. SET UP PARAM GRID
# ─────────────────────────────────────────────────────────────────────────────
param_grid = {
    'model__n_estimators':    [400, 800, 1200],
    'model__max_depth':       [6,10,14],
    'model__learning_rate':   [0.01,0.05,0.1,0.2]
}

# ─────────────────────────────────────────────────────────────────────────────
# 5. GRIDSEARCHCV with 5-fold CV
# ─────────────────────────────────────────────────────────────────────────────
cv = KFold(n_splits=5, shuffle=True, random_state=random_state)

grid = GridSearchCV(
    estimator  = pipe,
    param_grid = param_grid,
    cv         = cv,
    scoring    = 'neg_root_mean_squared_error',
    n_jobs     = -1,
    verbose    = 2,
)

# ─────────────────────────────────────────────────────────────────────────────
# 6. MLflow RUN
# ─────────────────────────────────────────────────────────────────────────────
with mlflow.start_run():
    start_time = time.time()
    # log basic run params
    mlflow.log_param("test_size",    test_size)
    mlflow.log_param("model_name",  "XGBRegressor")

    logging.info("Starting GridSearchCV …")
    grid.fit(X_train, y_train)
    logging.info("GridSearchCV complete.")

    # # log every CV combo’s mean & std RMSE
    # cv_res = grid.cv_results_
    # for i, (mean_score, std_score, params) in enumerate(
    #     zip(cv_res['mean_test_score'], cv_res['std_test_score'], cv_res['params'])
    # ):
    #     mean_rmse = -mean_score
    #     mlflow.log_metric(f"cv_{i}_mean_rmse", mean_rmse)
    #     mlflow.log_metric(f"cv_{i}_std_rmse",  std_score)
    #     # also log the param values for this combo
    #     for p_name, p_val in params.items():
    #         mlflow.log_param(f"cv_{i}_{p_name}", p_val)

    # best CV results
    best_rmse   = -grid.best_score_
    best_params =  grid.best_params_
    mlflow.log_metric("cv_best_rmse", best_rmse)
    for p_name, p_val in best_params.items():
        mlflow.log_param(f"best_{p_name}", p_val)

    logging.info(f"Best CV RMSE = {best_rmse:.4f}")
    logging.info(f"Best params   = {best_params}")

    # ─────────────────────────────────────────────────────────────────────────
    # 7. EVALUATE ON HELD-OUT TEST SET
    # ─────────────────────────────────────────────────────────────────────────
    logging.info("Evaluating best model on test set …")
    best_model = grid.best_estimator_
    y_pred     = best_model.predict(X_test)

    r2   = r2_score(y_test, y_pred)
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    mlflow.log_metric("test_r2_score",                r2)
    mlflow.log_metric("test_mean_absolute_error",     mae)
    mlflow.log_metric("test_root_mean_squared_error", rmse)

    logging.info(f"Test R²   = {r2:.4f}")
    logging.info(f"Test MAE  = {mae:.4f}")
    logging.info(f"Test RMSE = {rmse:.4f}")

    # ─────────────────────────────────────────────────────────────────────────
    # 8. SAVE & LOG FINAL MODEL
    # ─────────────────────────────────────────────────────────────────────────
    model_dir = "xgb_cv_model"
    if os.path.exists(model_dir):
        shutil.rmtree(model_dir)
    mlflow.sklearn.save_model(best_model, path=model_dir)
    mlflow.log_artifacts(model_dir, artifact_path="model")

    elapsed = time.time() - start_time
    logging.info(f"Run completed in {elapsed:.1f}s")
```
